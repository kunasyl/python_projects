```python
import requests
from bs4 import BeautifulSoup as bs
import csv
import pandas as pd
```

```python
from time import sleep
from random import randint
```

```python
headers = {
    'accept': '*/*',
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'
}
base_url = 'https://www.kinopoisk.ru/top/navigator/order/rating/page/1/#results'
```

```python
flats = []
urls = []
urls.append(base_url)
session = requests.Session()
request = session.get(base_url, headers=headers)
if request.status_code == 200:
    print('successful connection')
else:
    print('ERROR')
```

```python
try:
    page_counter = 471
    for i in range(page_counter):
        url = f'https://www.kinopoisk.ru/top/navigator/order/rating/page/{i+1}/#results'
        if url not in urls:
            urls.append(url)
except:
    print('err')
    pass

for url in urls:
    flats_len = 0
    request = session.get(url, headers=headers)
    soup = bs(request.content, 'lxml')
    divs = soup.find_all('div', attrs={'class': 'item'})
    for div in divs:
        name = div.find('div', attrs={'class': 'name'}).text

        info = div.find('div', attrs={'class': 'name'}).contents[2]
        info = [str for str in info.stripped_strings]
        title = info[0]
        if len(info) < 2:
            duration = None
        else:
            duration = info[1]

        country = div.find_all('span', attrs={'class': 'gray_text'})[0].contents[0].strip().strip(',...')

        director = div.find_all('span', attrs={'class': 'gray_text'})[0].i
        if director is None:
            director = None
        else:
            director = director.text

        genre = div.find_all('span', attrs={'class': 'gray_text'})[0].contents[-1].split()

        cast = div.find_all('span', attrs={'class': 'gray_text'})[1].text

        kp_rating = div.find('div', attrs={'class': 'numVote'}).text[:5]
        kp_votes_num = div.find('div', attrs={'class': 'numVote'}).text[7:-1]

        imdb = div.find('div', attrs={'class': 'imdb'})
        if imdb is not None:
            imdb_rating = imdb.text[:10]
            imdb_votes_num = imdb.text[10:]
        else:
            imdb_rating = None
            imdb_votes_num = None

        flats.append({
            'name': name,
            'title': title,
            'duration': duration,
            'country': country,
            'director': director,
            'genre': genre,
            'cast': cast,
            'kp_rating': kp_rating,
            'kp_votes_num': kp_votes_num,
            'imdb_rating': imdb_rating,
            'imdb_votes_num': imdb_votes_num
        })

    print(url)
    print(len(flats))
    sleep(randint(5,10))
```

```python
def files_writer(flats):
    with open(r"kinopoisk.csv", "w", encoding='utf-8') as file:
        a_pen = csv.writer(file)
        a_pen.writerow(('name', 'title', 'duration', 'country', 'director', 'genre', 'cast',
                        'kp_rating', 'kp_votes_num', 'imdb_rating', 'imdb_votes_num'))
        for flat in flats:
            a_pen.writerow((flat['name'], flat['title'], flat['duration'], flat['country'], flat['director'], flat['genre'],
                            flat['cast'], flat['kp_rating'], flat['kp_votes_num'], flat['imdb_rating'], flat['imdb_votes_num']))
```

```python
files_writer(flats)
```
